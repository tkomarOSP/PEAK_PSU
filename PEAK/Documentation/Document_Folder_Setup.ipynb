```python
from __future__ import annotations

from dataclasses import dataclass
from pathlib import Path
from datetime import datetime
import textwrap
import yaml  # pip install pyyaml (often already available in many Jupyter envs)


# -----------------------------
# Config you may want to change
# -----------------------------
DEFAULT_PROJECT_NAME = "My_System_Project"
DEFAULT_OUTPUT_DIR = Path.cwd() / "csid_arch_docs"


@dataclass(frozen=True)
class Section:
    """Represents a document section as a folder with a README.md."""
    key: str           # stable ID for agents and scripts
    folder: str        # folder name (ordered)
    title: str         # display title
    readme: str        # README.md content


def _md(title: str, body: str) -> str:
    """Small helper to produce consistent markdown."""
    return f"# {title}\n\n{body.strip()}\n"


def build_sections() -> list[Section]:
    """Defines the folder structure that mirrors the lesson document."""
    return [
        Section(
            key="roadmap",
            folder="01_Roadmap",
            title="Roadmap",
            readme=_md(
                "Roadmap",
                textwrap.dedent("""
                **Purpose:** Explain how this architecture package is organized and how the work will be developed over time.

                ## Include
                - Project context (what problem are we solving?)
                - Architecture development approach (phases, iterations)
                - Deliverables and milestones
                - Assumptions and constraints driving the plan

                ## Prompts
                - What’s the minimum viable architecture description we can deliver in 2 weeks?
                - What must be modeled vs documented in prose?
                - What are the planned checkpoints for review (stakeholders, instructors, peers)?

                ## “Good looks like”
                - Clear sequencing, not just a list
                - Explicit assumptions and dependencies
                - Review points mapped to stakeholders
                """)
            )
        ),
        Section(
            key="system_overview",
            folder="02_System_Overview",
            title="System Overview",
            readme=_md(
                "System Overview",
                textwrap.dedent("""
                **Purpose:** Give readers enough context to understand the system before reading views.

                ## Include
                - System mission / objectives
                - Scope and boundary
                - Stakeholders and their concerns
                - Operational concept summary (high level)
                - Architectural drivers (quality attributes, constraints)

                ## Suggested artifacts (optional)
                - context diagram (image)
                - stakeholder table
                - glossary

                ## “Good looks like”
                - A new reader can explain the system in 60 seconds
                - Clear boundary: what’s in vs out
                - Drivers are concrete (not vague aspirations)
                """)
            )
        ),
        Section(
            key="view_template",
            folder="03_View_Template",
            title="Architecture View Template",
            readme=_md(
                "Architecture View Template",
                textwrap.dedent("""
                **Purpose:** Define how each view will be documented so the package stays consistent.

                ## Include
                - Standard fields for a view:
                  - Name / viewpoint
                  - Stakeholder concerns addressed
                  - Modeling notation/tools used
                  - Key elements and relationships
                  - Rationale and assumptions
                  - Traceability links (requirements ↔ architecture)
                - Definition of “done” for a view

                ## Prompts
                - What does every view need so another team could reuse it?
                - What is the “contract” between prose and model diagrams?

                ## “Good looks like”
                - Repeatable and checklist-like
                - Encourages traceability and review
                """)
            )
        ),
        Section(
            key="views",
            folder="04_Views",
            title="Architecture Views",
            readme=_md(
                "Architecture Views",
                textwrap.dedent("""
                **Purpose:** Organize the core architectural descriptions by view type.

                This folder contains subfolders for:
                - Functional View
                - Logical View
                - Physical View

                Each subfolder should follow the Architecture View Template.
                """)
            )
        ),
        Section(
            key="functional_view",
            folder="04_Views/01_Functional_View",
            title="Functional View",
            readme=_md(
                "Functional View",
                textwrap.dedent("""
                **Purpose:** Describe what the system must do—independent of implementation.

                ## Include
                - Key functions and decomposition
                - Functional interfaces / exchanges
                - Functional scenarios / chains
                - Allocation hints (optional): functions to logical elements

                ## Prompts
                - What are the top 5 mission functions?
                - What functional sequences realize a primary operational scenario?
                - Where are the major control points and information flows?

                ## “Good looks like”
                - Function names are action-oriented and testable
                - Scenarios connect to stakeholder needs
                """)
            )
        ),
        Section(
            key="logical_view",
            folder="04_Views/02_Logical_View",
            title="Logical View",
            readme=_md(
                "Logical View",
                textwrap.dedent("""
                **Purpose:** Describe the logical decomposition and interactions that realize the functions.

                ## Include
                - Logical components / subsystems
                - Logical interfaces and exchanges
                - Allocation: functions → logical components
                - Key interaction patterns (message/event/data flow)

                ## Prompts
                - What are the main logical responsibilities?
                - What interactions must exist regardless of physical tech choice?
                - Where are the major performance, safety, security drivers?

                ## “Good looks like”
                - Coherent responsibilities (low coupling, high cohesion)
                - Clear interface definitions
                """)
            )
        ),
        Section(
            key="physical_view",
            folder="04_Views/03_Physical_View",
            title="Physical View",
            readme=_md(
                "Physical View",
                textwrap.dedent("""
                **Purpose:** Describe the physical realization (hardware/software deployment, network, etc.).

                ## Include
                - Physical nodes (devices, ECUs, servers, sensors, actuators)
                - Deployment / allocation: logical → physical
                - Physical interfaces (buses, networks, connectors)
                - Constraints (mass/power/latency/environment)

                ## Prompts
                - What are the candidate physical platforms?
                - What constraints dominate (power, weight, connectivity, cost)?
                - What are the key deployment scenarios?

                ## “Good looks like”
                - Clear mapping from logical to physical
                - Constraints stated with units/limits where possible
                """)
            )
        ),
        Section(
            key="view_mappings",
            folder="05_Mapping_Between_Views",
            title="Mapping Between Views",
            readme=_md(
                "Mapping Between Views",
                textwrap.dedent("""
                **Purpose:** Make traceability explicit across views (and to requirements).

                ## Include
                - Function → Logical allocation table
                - Logical → Physical allocation table
                - Requirements → architectural element traces (if available)
                - Gaps / TBDs explicitly listed

                ## Prompts
                - What functions are not allocated yet?
                - What logical components have no physical realization yet?
                - What requirements have no architectural anchor?

                ## “Good looks like”
                - Tables are complete enough to review
                - TBDs are visible (not hidden)
                """)
            )
        ),
        Section(
            key="rationale",
            folder="06_Rationale",
            title="Rationale",
            readme=_md(
                "Rationale",
                textwrap.dedent("""
                **Purpose:** Justify key decisions and tradeoffs; capture why the architecture looks this way.

                ## Include
                - Architectural decisions (ADRs) or decision log
                - Trade studies (even lightweight)
                - Risks and mitigations
                - Alternatives considered and why rejected
                - Impact on quality attributes

                ## Prompts
                - What decision, if wrong, would hurt us most?
                - What constraints forced a design choice?
                - Where do we need validation/experiments?

                ## “Good looks like”
                - Decision rationale ties to drivers
                - Alternatives are documented (prevents rework later)
                """)
            )
        ),
        Section(
            key="appendices",
            folder="07_Appendices",
            title="Appendices",
            readme=_md(
                "Appendices",
                textwrap.dedent("""
                **Purpose:** Hold supporting material that shouldn’t clutter the main flow.

                ## Include
                - Glossary
                - References
                - Diagrams exports
                - Data dictionaries
                - Assumption log

                ## “Good looks like”
                - Cleanly separated support materials
                - Links back to the main sections
                """)
            )
        ),
    ]


def scaffold_project(project_name: str,
                     output_dir: Path = DEFAULT_OUTPUT_DIR,
                     overwrite: bool = False) -> Path:
    """
    Create the folder structure and README.md files.

    Returns the path to the project root created.
    """
    timestamp = datetime.utcnow().strftime("%Y%m%dT%H%M%SZ")
    project_root = output_dir / f"{project_name}_ARCH_{timestamp}"

    if project_root.exists():
        if not overwrite:
            raise FileExistsError(f"Project folder already exists: {project_root}")
    project_root.mkdir(parents=True, exist_ok=True)

    sections = build_sections()

    # Create each folder + README.md
    for s in sections:
        folder_path = project_root / s.folder
        folder_path.mkdir(parents=True, exist_ok=True)
        readme_path = folder_path / "README.md"
        readme_path.write_text(s.readme, encoding="utf-8")

    # Top-level README
    top_readme = _md(
        f"{project_name} — Architecture Package",
        textwrap.dedent(f"""
        This repository/folder is a scaffold for an architecture description package.

        ## How to use
        1. Work section-by-section in order.
        2. Place diagrams, exports, and tables in the relevant folders.
        3. Keep artifacts small and reviewable (good for Git diffs).
        4. If you use agents/tools, point them at `manifest.yaml` to find targets.

        ## Suggested conventions
        - Put images in `07_Appendices/` or alongside the view they support.
        - Prefer Markdown for prose so diffs are readable.
        - Add new decision records in `06_Rationale/` (e.g., ADR_001.md).

        Created: {timestamp} (UTC)
        """)
    )
    (project_root / "README.md").write_text(top_readme, encoding="utf-8")

    # Manifest for agents (simple index)
    manifest = {
        "project": project_name,
        "created_utc": timestamp,
        $"project_path": ,
        "sections": [
            {
                "key": s.key,
                "title": s.title,
                "path": str(Path(s.folder)),
                "readme": str(Path(s.folder) / "README.md"),
            }
            for s in sections
        ],
        "suggested_agent_targets": [
            {"key": "system_overview", "drop_zone": "02_System_Overview/"},
            {"key": "functional_view", "drop_zone": "04_Views/01_Functional_View/"},
            {"key": "logical_view", "drop_zone": "04_Views/02_Logical_View/"},
            {"key": "physical_view", "drop_zone": "04_Views/03_Physical_View/"},
            {"key": "view_mappings", "drop_zone": "05_Mapping_Between_Views/"},
            {"key": "rationale", "drop_zone": "06_Rationale/"},
        ],
    }
    (project_root / "manifest.yaml").write_text(yaml.safe_dump(manifest, sort_keys=False), encoding="utf-8")

    return project_root


# -----------------------------
# Run the scaffold
# -----------------------------
project_root = scaffold_project(DEFAULT_PROJECT_NAME)
project_root
```
